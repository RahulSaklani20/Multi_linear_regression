```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
from statsmodels.graphics.regressionplots import influence_plot
```

```python
data=pd.read_csv('50_Startups.csv')
data
```

```python
data.info()
```

```python
data1=data.rename({'R&D Spend':'RDS','Administration':'ADMS','Marketing Spend':'MKSP'},axis=1)
data1
```

### Correlation Analysis

```python
data1.corr()
```

```python
sns.set_style(style='darkgrid')
sns.pairplot(data1)
```

```python
model=smf.ols('Profit~RDS+ADMS+MKSP',data=data1).fit()
```

```python
model.params
```

```python
model.tvalues,np.round(model.pvalues,5)
```

```python
model.rsquared,model.rsquared_adj
```

### Collinearity check

```python

rsq_r=smf.ols("RDS~ADMS+MKSP",data=data1).fit().rsquared
vif_r=1/(1-rsq_r)

rsq_a=smf.ols("ADMS~RDS+MKSP",data=data1).fit().rsquared
vif_a=1/(1-rsq_a)

rsq_m=smf.ols("MKSP~RDS+ADMS",data=data1).fit().rsquared
vif_m=1/(1-rsq_m)

# Putting the values in Dataframe format
d1={'Variables':['RDS','ADMS','MKSP'],'Vif':[vif_r,vif_a,vif_m]}
Vif_df=pd.DataFrame(d1)
Vif_df
```

### Cooks Distance

```python
(c,_)=model.get_influence().cooks_distance
c
```

```python
fig=plt.figure(figsize=(20,7))
plt.stem(np.arange(len(data1)),np.round(c,5))
plt.xlabel('Row Index')
plt.ylabel('Cooks Distance')
plt.show()
```

```python
np.argmax(c) , np.max(c)
```
